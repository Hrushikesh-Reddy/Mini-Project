```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Algerian_forest_fires_dataset.csv', header=1)
data.head()
```

```python
data.info()
```

```python
data[data.isnull().any(axis=1)]
```

```python
data.loc[:122,'Region']=1
data.loc[122:,'Region']=2
data[['Region']] = data[['Region']].astype(int)
```

```python
data.isnull().sum()
```

```python
data = data.dropna().reset_index(drop=True)
data.shape
```

```python
data.loc[122]
```

```python
data = data.drop(122).reset_index(drop=True)
```

```python
data.columns
```

```python
data.columns = data.columns.str.strip()
```

```python
data[['day', 'month', 'year', 'Temperature','RH', 'Ws']] = data[['day', 'month', 'year', 'Temperature','RH', 'Ws']].astype(int)
```

```python
obj = [cols for cols in data.columns if data[cols].dtypes=='O']
for i in obj:
    if i != 'Classes':
        data[i] = data[i].astype(float)
```

```python
data.Classes = data.Classes.str.strip()
data.Classes.value_counts()
```
